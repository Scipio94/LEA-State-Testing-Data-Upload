## NJSLA Import Script
Creating a script that creates a template for NJSLA testing imports

```python
# importing packages
import pandas as pd
import datetime as dt
import string as str 
```

```python
# importing data
df = pd.read_excel(r"S:\ASSESSMENTS\NJSLA 2025\NJSLA 2025 ELA Math Summative Record File.xlsx")
```

```python
# adding test column to the df
df['TEST'] = 'NJSLA'
```

```python
# data cleaning
df['TEST DATE'] = df['Unit1OnlineTestStartDateTime'].dt.strftime('%m/%d/%Y')
```

```python
df['TestCode'].sort_values().unique()
```

```python
def MATHTEST(row):
    if row['TestCode'] in ['ALG01', 'ALG02','GEO01', 'MAT03', 
                           'MAT04', 'MAT05', 'MAT06','MAT07', 'MAT08']:
        return row['TestCode']
    else:
        pd.NA
df['MATHTEST'] = df.apply(MATHTEST, axis = 1)
```

```python
def ELATEST(row):
    if row['TestCode'] in ['ELA03', 'ELA04', 'ELA05', 'ELA06',
                           'ELA07','ELA08', 'ELA09']:
        return row['TestCode']
    else:
        pd.NA
df['ELATEST'] = df.apply(ELATEST, axis = 1)
```

```python
def MATHSCORE(row):
    if row['TestCode']in ['ALG01', 'ALG02','GEO01', 'MAT03', 
                           'MAT04', 'MAT05', 'MAT06','MAT07', 'MAT08']:
        return row['TestScaleScore']
    else:
        return pd.NA
df['MATHSCORE'] = df.apply(MATHSCORE, axis = 1)
```

```python
def ELASCORE(row):
    if row['TestCode']in ['ELA03', 'ELA04', 'ELA05', 'ELA06',
                           'ELA07','ELA08', 'ELA09']:
        return row['TestScaleScore']
    else:
        return pd.NA
df['ELASCORE'] = df.apply(ELASCORE, axis = 1)
```

```python
def MATHPL(row):
    if row['TestCode']in ['ALG01', 'ALG02','GEO01', 'MAT03', 
                           'MAT04', 'MAT05', 'MAT06','MAT07', 'MAT08']:
        return row['TestPerformanceLevel']
    else:
        return pd.NA
df['MATHPL'] = df.apply(MATHPL, axis = 1)
```

```python
def ELAPL(row):
    if row['TestCode']in ['ELA03', 'ELA04', 'ELA05', 'ELA06',
                           'ELA07','ELA08', 'ELA09']:
        return row['TestPerformanceLevel']
    else:
        return pd.NA
df['ELAPL'] = df.apply(ELAPL, axis = 1)
```

```python
df['AccountableSchoolCode'].unique()
```

```python
# entity formula
def entity(row):
    if row['AccountableSchoolCode'] == 50:
        return '010'
    elif row['AccountableSchoolCode'] == 60:
        return '008'
    elif row['AccountableSchoolCode'] == 70:
        return '009'
    elif row['AccountableSchoolCode'] == 85:
        return '002'
    elif row['AccountableSchoolCode'] == 90:
        return '004'
    elif row['AccountableSchoolCode'] == 100:
        return '005'
    elif row['AccountableSchoolCode'] == 110:
        return '006'
    elif row['AccountableSchoolCode'] == 120:
        return '003'
    elif row['AccountableSchoolCode'] == 130:
        return '007'
df['Entity'] = df.apply(entity, axis = 1)
df['Entity'] = df['Entity'].astype('str')
```

```python
df_math = df[['StateStudentIdentifier','LocalStudentIdentifier','LastOrSurname','FirstName',
   'GradeLevelWhenAssessed','TEST','TEST DATE','MATHTEST','MATHSCORE','MATHPL','Entity']]
    

df_math = df_math[df_math['MATHSCORE'].notna()]
```

```python
df_ela = df[['StateStudentIdentifier','LocalStudentIdentifier','LastOrSurname','FirstName',
             'ELATEST','ELASCORE','ELAPL', 'Entity' ]]
    

df_ela = df_ela[df_ela['ELASCORE'].notna()]
```
